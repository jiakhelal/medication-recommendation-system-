```python
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report,
    top_k_accuracy_score
)
```

```python
models = {
    "svm": SVC(kernel="linear", probability=True),
    "RandomForestClassifier": RandomForestClassifier(
        n_estimators=100,
        random_state=42
    ),
    "GradientBoostingClassifier": GradientBoostingClassifier(
        n_estimators=100,
        random_state=42,
    )
}

labels = np.unique(y_train)


for model_name, model in models.items():

    print("\n" + "="*60)
    print(f"Training model: {model_name}")

    # Train
    model.fit(x_train, y_train)

    # Predictions
    y_pred = model.predict(x_test)

    # Top-1 Accuracy
    acc = accuracy_score(y_test, y_pred)
    print(f"Top-1 Accuracy for {model_name}: {acc:.4f}")

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    print(f"Confusion Matrix for {model_name}:\n{cm}")

    # Classification Report
    print(f"Classification Report for {model_name}:")
    print(classification_report(y_test, y_pred))

    # Top-3 Accuracy
    probs = model.predict_proba(x_test)

    top3_acc = top_k_accuracy_score(
        y_test,
        probs,
        k=3,
       labels=labels
    )

    print(f"Top-3 Accuracy for {model_name}: {top3_acc:.4f}")
```

```python
import ast

def helper(dis):
    # =======================
    # Description
    # =======================
    desc = description_df[description_df['Disease'] == dis]['Description']
    desc = " ".join(desc.values)

    # =======================
    # Precautions (FIXED)
    # =======================
    pre_df = precautions_df[precautions_df['Disease'] == dis][
        ['Precaution_1', 'Precaution_2', 'Precaution_3', 'Precaution_4']
    ]

    precautions = []
    if not pre_df.empty:
        precautions = pre_df.iloc[0].dropna().tolist()

    # =======================
    # Medications
    # =======================
    med_series = medications_df[medications_df['Disease'] == dis]['Medication']
    medications = []
    for row in med_series.values:
        medications.extend(ast.literal_eval(row))

    # =======================
    # Diet
    # =======================
    diet_series = diets_df[diets_df['Disease'] == dis]['Diet']
    diet = []
    for row in diet_series.values:
        diet.extend(ast.literal_eval(row))

    # =======================
    # Workout
    # =======================
    wrkout = workout_df[workout_df['disease'] == dis]['workout']
    workout = wrkout.values.tolist()

    return desc, precautions, medications, diet, workout



```

```python
def normalize_symptom(symptom):
    return symptom.strip().lower().replace(" ", "_")
```

```python
def get_top_3_predictions(patient_symptoms, top_k=3):

    input_vector = np.zeros(len(symptoms_list))
    valid_found = False
    unknown = []

    for symptom in patient_symptoms:
        symptom = symptom.strip().lower()
        if symptom in symptoms_list:
            idx = symptoms_list.index(symptom)
            input_vector[idx] = 1
            valid_found = True
        else:
            unknown.append(symptom)

    if not valid_found:
        return {
            "error": "Please enter symptoms, not disease names.",
            "unknown_symptoms": unknown
        }

    probs = svc.predict_proba([input_vector])[0]
    top_indices = np.argsort(probs)[-top_k:][::-1]

    results = []
    for idx in top_indices:
        disease = disease_dict[idx]
        confidence = round(probs[idx] * 100, 2)

        desc, pre, med, die, wrk = helper(disease)

        results.append({
            "disease": disease,
            "confidence": confidence,
            "description": desc,
            "precautions": pre,
            "medications": med,
            "diet": die,
            "workout": wrk
        })

    return results

```

```python
symptoms = input("Enter the symptoms: ")

user_symptoms = [
    s.strip().lower().replace("_", " ")
    for s in symptoms.split(",")
]

top_results = get_top_3_predictions(user_symptoms)

print("\n================ TOP 3 PREDICTIONS ================\n")

for i, res in enumerate(top_results, start=1):

    print(f"\n#{i} Disease: {res['disease']} ({res['confidence']}%)")

    # ✅ Confidence disclaimer (per disease)
    if res["confidence"] < 50:
        print("⚠️ Low confidence prediction. Please consult a medical professional.")

    print("\nDescription:")
    print(res["description"])

    print("\nPrecautions:")
    for p in res["precautions"]:
        print("-", p)

    print("\nMedications:")
    for m in res["medications"]:
        print("-", m)

    print("\nWorkout:")
    for w in res["workout"]:
        print("-", w)

    print("\nDiet:")
    for d in res["diet"]:
        print("-", d)

    print("\n" + "-" * 60)
```
